##**Importing necessary libraries and ignoring warnings(if any)**

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

##**Reading the dataset**

```python
df=pd.read_csv('/content/car.csv') #pandas function is used.
```

```python
#displaying the content/data in the given csv file.
df.head(10)
```

##**Analyzing the data -**

```python
#cheking for Null Values
df.isna().sum()
```

```python
#removing the Null data
df=df[~df['fuel_type'].isna()]
```

```python
#reseting the index after droping
df.reset_index(drop=True)
```

```python
#Removing non year from Year Column
df=df[df['year'].str.isnumeric()]
```

```python
# making year as numeric
df['year']=df['year'].astype(int)
```

```python
# changing "Ask for Price" rather removing them from price column

df=df[df['Price']!="Ask For Price"]
```

```python
# making price as int type
df['Price']=df['Price'].str.replace(',','').astype(int)
```

```python
# removing "," & "kms" from kms-driven column
df['kms_driven'] = df['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
# making kms_driven column as int type
df = df[df['kms_driven'].str.isnumeric()]
df['kms_driven']=df['kms_driven'].astype(int)
```

```python
# slicing the car name to 3 words only

df['name'] = df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df.head()
```

```python
df.info()
```

```python
#restetting the indexes again
df.reset_index(drop=True)
```

```python
#statistical analysis

df.describe()
```

```python
#outlier removal

df=df[df['Price']<6e6].reset_index(drop=True)
df
```

```python
#make separate input and output data colums

X = df.drop(columns='Price')
y = df['Price']
```

```python
# differentiate training and testing data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
y_train.head()
```

```python
x_train.head()
```

```python
#'name','company','fuel_type' columns are categorical so we need to make them numeric

from sklearn.preprocessing import OneHotEncoder, StandardScaler
df1=OneHotEncoder()
df1.fit(df[['name','company','fuel_type']])
```

```python
from sklearn.compose import make_column_transformer
column_trans= make_column_transformer((OneHotEncoder(categories=df1.categories_),['name','company','fuel_type']),
                                     remainder='passthrough')
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error,r2_score

import math
from sklearn.pipeline import make_pipeline
```

```python
#XGB regressor

xgbr=XGBRegressor()
pipe=make_pipeline(column_trans,xgbr)
pipe.fit(x_train,y_train)
y_pred_xgbr=pipe.predict(x_test)
r2=r2_score(y_test,y_pred_xgbr)
r2 
#checking r2 score
#r2 score should lie between 0 & 1. The more closer it is towards 1, 
#better is the accuracy of the Regression model
```

```python
#let's run XGBRegression for 100 times and store the highest scores

score=[]
for i in range(100):
    x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=i)
    xgbr=XGBRegressor()
    pipe=make_pipeline(column_trans,xgbr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    score.append(r2_score(y_test,y_pred))
```

```python
score[np.argmax(score)]
```

```python
#saving according to the best item

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=np.argmax(score))
xgbr=XGBRegressor()
pipe=make_pipeline(column_trans,xgbr)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
r2_score(y_test,y_pred)
```

##**Testing our model prediction -**
